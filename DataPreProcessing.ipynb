As we are doing clustering to cluster songs with similar mood/genre, we need features which only define the type of song. So in this data we don't need id_songs, name_song, popularity_songs, explicit, id_artists, release_date, key, time_signature, followers, genres, name_artists, popularity_artists.

```python
#Removing unnecessary columns
unnescessary_columns = ['id_songs', 'name_song', 'popularity_songs', 'explicit', 'id_artists', 'release_date', 'key', 'time_signature', 'followers', 'genres', 'name_artists', 'popularity_artists']
music_data.drop(unnescessary_columns, axis = 1, inplace=True)
```

```python
# mode column is also not required
music_data.drop('mode', axis = 1, inplace = True)
```

```python
#Outlier treatment for duration_ms
Q1 = music_data.duration_ms.quantile(0.05)
Q3 = music_data.duration_ms.quantile(0.95)
IQR = Q3 - Q1
music_data = music_data[(music_data.duration_ms >= Q1 - 1.5*IQR) & (music_data.duration_ms <= Q3 + 1.5*IQR)]
```

```python
#Outlier treatment for danceability
Q1 = music_data.danceability.quantile(0.05)
Q3 = music_data.danceability.quantile(0.95)
IQR = Q3 - Q1
music_data = music_data[(music_data.danceability >= Q1-1.5*IQR) & (music_data.danceability <= Q3+1.5*IQR)]

#Outlier treatment for loudness
Q1 = music_data.loudness.quantile(0.05)
Q3 = music_data.loudness.quantile(0.95)
IQR = Q3 - Q1
music_data = music_data[(music_data.loudness >= Q1-1.5*IQR) & (music_data.loudness <= Q3+1.5*IQR)]

#Outlier treatment for speechiness
Q1 = music_data.speechiness.quantile(0.05)
Q3 = music_data.speechiness.quantile(0.95)
IQR = Q3 - Q1
music_data = music_data[(music_data.speechiness >= Q1-1.5*IQR) & (music_data.speechiness <= Q3+1.5*IQR)]

#Outlier treatment for instrumentalness
Q1 = music_data.instrumentalness.quantile(0.05)
Q3 = music_data.instrumentalness.quantile(0.95)
IQR = Q3 - Q1
music_data = music_data[(music_data.instrumentalness >= Q1-1.5*IQR) & (music_data.instrumentalness <= Q3+1.5*IQR)]

#Outlier treatment for liveness
Q1 = music_data.liveness.quantile(0.05)
Q3 = music_data.liveness.quantile(0.95)
IQR = Q3 - Q1
music_data = music_data[(music_data.liveness >= Q1-1.5*IQR) & (music_data.liveness <= Q3+1.5*IQR)]

#Outlier treatment for tempo
Q1 = music_data.tempo.quantile(0.05)
Q3 = music_data.tempo.quantile(0.95)
IQR = Q3 - Q1
music_data = music_data[(music_data.tempo >= Q1-1.5*IQR) & (music_data.tempo <= Q3+1.5*IQR)]
```
